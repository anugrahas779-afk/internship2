```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.plot(x,y,color='green',linestyle='--',marker='o')
plt.title('My plot')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```

```python
x=[1,2,3]
y1=[1,4,9]
y2=[1,2,3]
plt.plot(x,y1,label='Squares')
plt.plot(x,y2,label='Linear')
plt.legend()
plt.show()
```

```python
x=[1,2,3]
y1=[1,4,9]
y2=[1,2,3]
plt.plot(x,y1,label='Squares',color='yellow')
plt.plot(x,y2,label='Linear',color='red')
plt.legend(loc='center')
plt.show()
```

```python
x=[1,2,3]
y1=[1,4,9]
y2=[1,2,3]
plt.plot(x,y1,label='Squares',color='brown',linestyle='--',marker='o')
plt.plot(x,y2,label='Linear',color='red',linestyle='--',marker='o')
plt.legend()
plt.show()
```

```python
x=[1,2,3]
y1=[1,4,9]
y2=[1,2,3]
plt.plot(x,y1,label='Squares',color='brown',linestyle='--',marker='')
plt.plot(x,y2,label='Linear',color='red',linestyle='--',marker='+')
plt.legend()
plt.show()
```

```python
x=[1,2,3]
y1=[1,4,9]
y2=[1,2,3]
plt.plot(x,y1,label='Squares',color='brown',linestyle='--',marker='*')
plt.plot(x,y2,label='Linear',color='red',linestyle='--',marker='*')
plt.legend(loc='lower left')
plt.show()
```

```python
categories=['A','B','C']
values=[5,7,3]
plt.bar(categories,values)
plt.show()
```

```python
categories=['A','B','C']
values=[5,7,3]
plt.bar(categories,values,color=['cyan','pink','blue'])
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values,color=['skyblue', 'orange', 'green'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Different Colors')
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values,color=['cyan', 'pink', 'darkblue'],width=1.0)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Different Colors')
plt.show()
```

```python
data=[1,1,2,3,3,3,4,4,5]
plt.hist(data,bins=5,color='pink')
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,25,30]
plt.scatter(x,y,marker='*',color='darkblue')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [1, 2, 3])
plt.subplot(1, 2, 2)
plt.plot([1, 2, 3], [3, 2, 1])
plt.show()
```
