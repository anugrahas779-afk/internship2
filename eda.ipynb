```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.histplot(df['sepal length (cm)'], kde=True, ax=axes[0, 0], color='green')
axes[0, 0].set_title('Sepal Length Distribution')
sns.histplot(df['sepal width (cm)'], kde=True, ax=axes[0, 1], color='skyblue')
axes[0, 1].set_title('Sepal Width Distribution')
sns.histplot(df['petal length (cm)'], kde=True, ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Petal Length Distribution')
sns.histplot(df['petal width (cm)'], kde=True, ax=axes[1, 1], color='red')
axes[1, 1].set_title('Petal Width Distribution')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop('Iris type', axis=1))
plt.title('Boxplot of All Numerical Features')
plt.ylabel('Centimeters (cm)')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
species = df['Iris type'].unique()
colors = ['red', 'green', 'blue']
for i in range(len(species)):
    x = df[df['Iris type'] == species[i]]
    plt.scatter(x['sepal length (cm)'],x['sepal width (cm)'],color=colors[i],label=species[i])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Width by Iris Type')
plt.legend(title='Iris Type')
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("Iris_Data.csv")
```

```python
print(type(df))
print(df.columns)
```

```python
iris = df.rename(columns={
    'sepal length (cm)': 'Sepal.Length',
    'sepal width (cm)': 'Sepal.Width',
    'petal length (cm)': 'Petal.Length',
    'petal width (cm)': 'Petal.Width',
    'Iris type': 'species'
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate correlation matrix
corr = df.drop('Iris type', axis=1).corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt=".2f")

plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.boxplot(x='species', y='sepal_length', data=iris, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Sepal Length by Species')
sns.boxplot(x='species', y='sepal_width', data=iris, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Sepal Width by Species')
sns.boxplot(x='species', y='petal_length', data=iris, ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Petal Length by Species')
sns.boxplot(x='species', y='petal_width', data=iris, ax=axes[1, 1], palette='Set2')
axes[1, 1].set_title('Petal Width by Species')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
sns.boxplot(x='Iris type', y='sepal length (cm)', data=df, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Sepal Length by Iris Type')
sns.boxplot(x='Iris type', y='sepal width (cm)', data=df, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Sepal Width by Iris Type')
sns.boxplot(x='Iris type', y='petal length (cm)', data=df, ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Petal Length by Iris Type')
sns.boxplot(x='Iris type', y='petal width (cm)', data=df, ax=axes[1, 1], palette='Set2')
axes[1, 1].set_title('Petal Width by Iris Type')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns.violinplot(
    x='species',
    y='Petal.Length',data=iris,inner='quartile',palette='pastel')
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
from pandas.plotting import andrews_curves
plt.figure(figsize=(10, 6))
andrews_curves(iris, 'species', colormap='viridis')
plt.title('Andrews Curves for Iris Dataset')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
plt.figure(figsize=(10, 6))
parallel_coordinates(iris, 'species', colormap='viridis')
plt.title('Parallel Coordinates Plot for Iris Dataset')
plt.show()
```
